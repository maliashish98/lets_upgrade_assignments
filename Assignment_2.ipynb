Q1. Creat an Empty list. Accept 10 numbers from the user and append to it the list if it is an even number.


```
my_list =[]
for i in range(0,10):
  ele = int(input())
  if(ele % 2 == 0):
    my_list.append(ele)

print(my_list)
```

Q4.The trace of robot movement is shown as the following:
UP 5
DOWN 3
LEFT 3
RIGHT 2
Then, the output of the program should be:
2
The numbers after the direction are steps.

```
pos = {"x":0,"y":0}
num = int(input())

for _ in range (num):
    command =  input().split(" ")      # ACCEPT MOVEMENT COMMAND AND STORE AS A LIST
    if command[0] == "UP":             # EXTRACT DIRECTION AND COMPARE
        pos["y"] += int(command[1])    # INCREMENT/DECREMENT APPROPRIATE CO-ORDINATES
    if command[0] == "DOWN":
        pos["y"] -= int(command[1])
    if command[0] == "LEFT":
        pos["x"] -= int(command[1])
    if command[0] == "RIGHT":
        pos["x"] += int(command[1])

print(int(round((pos["x"]**2 + pos["y"]**2)**0.5)))   # DISTANCE FROM ORIGIN
```
